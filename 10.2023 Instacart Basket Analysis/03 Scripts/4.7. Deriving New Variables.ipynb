# 3. Defining new functions

```python
# define a function

def price_label(row):
    
    if row['prices'] <=5:
        return 'low range product'
    elif(row['prices'] > 5) and (row['prices'] <= 15):
        return 'mid range product'
    elif row['prices'] > 15:
        return 'High range product'
    else:return np.nan
```

```python
# apply the function

df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
# use the max option to check out the most expensive product within the dataset
df['prices'].max()
```

```python
# use loop function to assign different values in new column
result = []

for value in df_ords_prods_merged["orders_day_of_week"]:
  if value == 0 or value == 1:  # 0 and 1 correspond to the busiest days
    result.append("Busiest days")
  elif value == 3 or value == 4:  # 3 and 4 correspond to the slowest days
    result.append("Slowest days")
  else:
    result.append("Regularly busy")
```

```python
# create a new column within the dataframe to combine new labels
df_ords_prods_merged['busiest_slowest_days'].value_counts(dropna = False)
```

```python
# to start we need to know on which hour the most orders take place
df_ords_prods_merged['order_hour'].value_counts(dropna = False)
```

"Most orders" for the top 25% of hours (hours with the highest counts),
"Fewest orders" for the bottom 25% of hours (hours with the lowest counts),
"Average orders" for the remaining 50%


Here are the hours that fall into each category:
    Most orders: 10 AM, 11 AM, 12 PM, 1 PM, 2 PM, 3 PM 
    Fewest orders: 12 AM, 1 AM, 2 AM, 3 AM, 4 AM, 5 AM 

```python
# Create a for-loop if statement using the above breakdown
label_order_period = []

for value in df_ords_prods_merged["order_hour"]:
  if value in [10, 11, 13, 14, 15, 12]:
    label_order_period.append("Most orders")
  elif value in [0, 1, 2, 3, 4, 5]:
    label_order_period.append("Fewest orders")
  else:
    label_order_period.append("Average orders")
```

```python
# Apply the function to each hour in the DataFrame
df_ords_prods_merged['busiest_period_of_day'] = label_order_period
```

```python
# check the frequency for the new column busiest hour of day
df_ords_prods_merged['busiest_period_of_day'].value_counts(dropna = False)
```

"Most orders" have the highest number of orders, which aligns with the expectation that these time slots would be the busiest in the day.

The "Average orders" category is very close to the "Most orders" count, suggesting that the distribution of orders outside the peak times is still quite high and relatively even throughout the day.

The low number for "Fewest orders" suggests that there is a stark contrast in order volumes during these hours compared to peak times.

with these insights, Instacart could use this data to manage their resources more effectively, scheduling maintenance or updates during the "Fewest orders" periods
